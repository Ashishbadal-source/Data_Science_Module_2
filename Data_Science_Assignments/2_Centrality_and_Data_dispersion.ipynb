```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("diabetes.csv")
```

```python
data.head()
```

```python
median_val = data["SkinThickness"].median()
data["SkinThickness"] = data["SkinThickness"].replace(0, median_val)

median_val = data["Insulin"].median()
data["Insulin"] = data["Insulin"].replace(0, median_val)

median_val = data["BloodPressure"].median()
data["BloodPressure"] = data["BloodPressure"].replace(0, median_val)
```

```python
mean_age = data["Age"].mean()
mean_glucose = data["Glucose"].mean()
mean_BP = data["BloodPressure"].mean()
mean_BMI = data["BMI"].mean()
mean_PdF = data["DiabetesPedigreeFunction"].mean()
```

```python
median_age = data["Age"].median()
median_glucose = data["Glucose"].median()
median_BP = data["BloodPressure"].median()
median_BMI = data["BMI"].median()
median_PdF = data["DiabetesPedigreeFunction"].median()
```

```python
mode_age = data["Age"].mode()
mode_glucose = data["Glucose"].mode()
mode_BP = data["BloodPressure"].mode()
mode_BMI = data["BMI"].mode()
mode_PdF = data["DiabetesPedigreeFunction"].mode()
```

```python
max_age = data["Age"].max()
max_glucose = data["Glucose"].max()
max_BP = data["BloodPressure"].max()
max_BMI = data["BMI"].max()
max_PdF = data["DiabetesPedigreeFunction"].max()
```

```python
min_age = data["Age"].min()
min_glucose = data["Glucose"].min()
min_BP = data["BloodPressure"].min()
min_BMI = data["BMI"].min()
min_PdF = data["DiabetesPedigreeFunction"].min()
```

```python
midRange_age = (min_age + max_age)/2
midRange_glucose = (min_glucose + max_glucose)/2
midRange_BP = (min_BP + max_BP)/2
midRange_BMI = (min_BMI + max_BMI)/2
midRange_PdF = (min_PdF + max_PdF)/2
```

```python
print("MID RANGE AGE :" ,midRange_age)
print("MID RANGE GLUCOSE : ",midRange_glucose)
print("MID RANGE BP :" ,midRange_BP)
print("MID RANGE BMI :" ,midRange_BMI)
print("MID RANGE DPF:" ,midRange_PdF )
```
