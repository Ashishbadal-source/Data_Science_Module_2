```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("movie_data.csv")
numeric_df = df.select_dtypes(include=["float64", "int64", "bool"])
corr = numeric_df.corr()
top_corr_features = corr["amount_paid"].drop("amount_paid").abs().sort_values(ascending=False).head(3).index.tolist()

focused_corr = numeric_df[["amount_paid"] + top_corr_features].corr() 

plt.figure(figsize=(6, 4))
sns.heatmap(focused_corr, annot=True, cmap="coolwarm", fmt=".2f", cbar=True)
plt.title("Focused Correlation Heatmap: amount_paid vs Top Features")
plt.show()

```
